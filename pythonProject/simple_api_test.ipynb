```python
import time


def load_greed_fear(iso_date,db_engine):
   url = 'https://production.dataviz.cnn.io/index/fearandgreed/graphdata/' + iso_date
   print(url)
   headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
   }
   response = requests.get(url, headers=headers)
   content = response.json()
   df = pd.json_normalize(content, max_level=0)
   df = df.applymap(lambda v: v['score'])
   df['Date'] = date.fromisoformat(iso_date)   
   df.to_sql('CNN_GREED_FEAR', db_engine, if_exists='append')


def load_cnn_period(start_date, end_date):
   start_date = date.fromisoformat(start_date)
   end_date = date.fromisoformat(end_date)
   db_engine = sqlalchemy.create_engine('sqlite:///D:\\sqlite\\data\\finance.db')
   while start_date < end_date:
      weekday = start_date.weekday()
      if weekday < 5:
         load_greed_fear(start_date.isoformat(),db_engine)
      start_date = start_date + timedelta(days=1)
   db_engine.dispose()
          
        
        
      
   
```
